```python
import pandas as pd
df = pd.read_csv("orders.csv", encoding='latin1',error_bad_lines=False, na_values=['Not Available', 'unknown'])
```

```python
df.head(4)
```

```python
df['Ship Mode'].unique()
```

```python
#rename column name- make it lower case and reeplace spacee with underscore
# df.rename{'Order Id':'order_id','Order Date':'order_date','Ship Mode':'ship_mode','Segment':'segment','Country':'country','City':'city','State':'state','Postal Code':'postal_code','Region':'region','Category':'category','Sub Category':'sub_category','Product Id':'product_id','cost price':'cost_price','List Price':'list_price'	,'Quantity':'quality','Discount Percent':'discount_price'}
df.columns=df.columns.str.lower()

df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#derive new column discount,sale_price and profit
df['discount'] = df['list_price']*df['discount_percent']/100
df['discount']
```

```python
#sale_price = list_price-discount
df['sale_price']=df['list_price']-df['discount']
df['sale_price']
```

```python
#profit = sale_price-cost_price
df['Profit'] = df['sale_price']-df['cost_price']
```

```python
df['Profit']
```

```python
#convert order date from object data type to datetime - we use pandas datetime
# df.dtypes
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df
```

```python
#drop cost price ,list price , discount percentage
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
